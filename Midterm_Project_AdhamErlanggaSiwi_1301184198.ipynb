Lakukanlah _trend analysis_ nilai _daily percentage change_ dari ketiga harga saham tersebut selama tahun 2019, lalu tampilkan dan jelaskan melalui visualisasi data.

#### Code

###### Ekplorasi Data

```python
#import dataset
df_nas = pd.read_csv("stock_market/Nasdaq.csv")
df_hang = pd.read_csv("stock_market/Hang_Seng.csv")
df_nik = pd.read_csv("stock_market/Nikkei.csv")
```

```python
#pembulatan 2 angka di belakang koma
df_nas = df_nas.round(2)
df_hang = df_hang.round(2)
df_nik = df_nik.round(2)
```

```python
#melihat jumlah baris dan kolom dataset
print("Jumlah baris dan kolom dataset nasdaq:",df_nas.shape)
print("Jumlah baris dan kolom dataset hang seng:",df_hang.shape)
print("Jumlah baris dan kolom dataset nikkei:",df_nik.shape)
```

```python
#mengecek apakah ada nilai null pada dataset
print("Nasdaq")
print(df_nas.isnull().sum())
print("")
print("Hang Seng")
print(df_hang.isnull().sum())
print("")
print("Nikkei")
print(df_nik.isnull().sum())
```

```python
#Drop entri null pada dataset
df_hang.dropna(inplace = True, axis = 0)
df_nik.dropna(inplace = True, axis = 0)
print("Hang Seng")
print(df_hang.isnull().sum())
print("")
print("Nikkei")
print(df_nik.isnull().sum())
```

```python
df_nas['Date'] = pd.to_datetime(df_nas['Date'])
df_hang['Date'] = pd.to_datetime(df_hang['Date'])
df_nik['Date'] = pd.to_datetime(df_nik['Date'])
```

```python
#set datetime sebagai index
df_nas.index = df_nas["Date"]
df_hang.index = df_hang["Date"]
df_nik.index = df_nik["Date"]
```

```python
#drop tahun selain tahun 2019
df_nas = df_nas.loc['2019-01-01':'2019-12-31']
df_hang = df_hang.loc['2019-01-01':'2019-12-31']
df_nik = df_nik.loc['2019-01-01':'2019-12-31']
```

```python
#membuat kolom baru Day_Perc_Change
df_nas["Day_Perc_Change"] = df_nas["Adj Close"].pct_change()*100
df_hang["Day_Perc_Change"] = df_hang["Adj Close"].pct_change()*100
df_nik["Day_Perc_Change"] = df_nik["Adj Close"].pct_change()*100
```

```python
df_nas.dropna(inplace = True, axis = 0)
df_hang.dropna(inplace = True, axis = 0)
df_nik.dropna(inplace = True, axis = 0)
```

###### Trend Analysis

```python
def trend(x):
    if x > -0.5 and x <= 0.5:
        return "Slight or No change"
    elif x > 0.5 and x <= 1:
        return "Slight Positive"
    elif x > -1 and x <= -0.5:
        return "Slight Negative"
    elif x > 1 and x <= 3:
        return "Positive"
    elif x > -3 and x <= -1:
        return "Negative"
    elif x > 3 and x <= 7:
        return "Among top gainers"
    elif x > -7 and x <= -3:
        return "Among top losers"
    elif x > 7:
        return "Bull run"
    elif x <= -7:
        return "Bear drop"

#membuat kolom baru untuk mencatat trend
df_nas["Trend"]= df_nas["Day_Perc_Change"].apply(lambda x:trend(x))
df_hang["Trend"]= df_hang["Day_Perc_Change"].apply(lambda x:trend(x))
df_nik["Trend"]= df_nik["Day_Perc_Change"].apply(lambda x:trend(x))
```

```python
df_nas['Trend'].unique()
```

###### Visualisasi

```python
fig, (ax1,ax2,ax3) = plt.subplots(1,3,figsize=(15,10))

pie_label = sorted([i for i in df_nas.loc[:, 'Trend'].unique()])

#Pie chart Nasdaq
nas_pie_data = df_nas.groupby('Trend')
ax1.pie(nas_pie_data['Trend'].count(), autopct = '%1.1f%%', radius = 1.3)
ax1.set_title("Nasdaq", y = 1.1, fontweight="bold")

#Pie chart Hangseng
hang_pie_data = df_hang.groupby('Trend')
ax2.pie(hang_pie_data['Trend'].count(), autopct = '%1.1f%%', radius = 1.3)
ax2.set_title("Hang Seng", y = 1.1, fontweight="bold")

#Pie chart Nikkei
nik_pie_data = df_nik.groupby('Trend')
ax3.pie(nik_pie_data['Trend'].count(), autopct = '%1.1f%%', radius = 1.3)
ax3.set_title("Nikkei", y = 1.1, fontweight="bold")


fig.suptitle("Trend Daily Percentage Change selama tahun 2019 dari ketiga perusahaan ", y = 0.77)
fig.legend(labels = pie_label, title = "Trend Legend", bbox_to_anchor=(1.03,0.8))
plt.show()
```

#### Penjelasan 
Tahapan yang dilakukan untuk trend analysis mula-mula ialah bagian eksplorasi data. 

Pada bagian eksplorasi data dilakukan import file berisi informasi perusahaan (dataset), lalu melihat jumlah data pada perusahaan dengan mengecek jumlah baris dan kolom pada dataset. setelah itu agar data dapat lebih mudah dibaca dilakukan pembulatan angka sebanyak 2 angka di belakang koma. kemudian pengecekan apakah ada nilai yang kosong pada tiap baris, bila ditemukan maka akan dibuang (drop) baris yang mengandung nilai kosong. Lalu mengubah tipe data kolom tanggal serta menjadikan kolom tanggal sebagai index dataset. Setelah itu mengambil data yang hanya tahun 2019. Kemudian, membuat kolom baru yang berisi selisih harga tutup antara 2 hari berturut-turut lalu drop baris yang memiliki nilai kosong pada kolom tersebut. Terakhir melakukan trend analysis pada data dengan membuat fungsi trend mengikuti kondisi trend yang ditentukan dari tabel, lalu membuat kolom trend yang diisi dengan menggunakan fungsi trend yang telah dibuat.

Lalu Memvisualisasikan data ketiga perusahaan dengan pie chart. melalui grafik pie dapat dilihat bahwa trend untuk Nasdaq mayoritasnya Slight negative. untuk perusahaan Hang Seng mayoritas trend Slight Positive dan untuk perusahaan Nikkei memiliki presentase mayoritas trend Slight Positive yang lebih baik dari Hang Seng

### Soal 4 (30 poin)
Lakukanlah analisis volatilitas dari ketiga harga saham tersebut selama tahun 2019, lalu tampilkan dan jelaskan hasil visualisasinya.

#### Code

```python
#Menggabungkan kolom Adj Close
df_adj = pd.concat([df_nas["Adj Close"], df_hang["Adj Close"], df_nik["Adj Close"]], axis = 1)
df_adj.columns = ["Nasdaq", "Hang Seng", "Nikkei"]

#reset index dari tanggal menjadi index biasa
df_adj = df_adj.reset_index()

#drop kolom tanggal
df_adj = df_adj.drop(["Date"], axis = 1)

#melihat dataset
df_adj.head()
```

```python
#drop baris yang mengandung nilai kosong
df_adj.dropna(inplace = True, axis = 0)

# menampilkan baris awal
df_adj.head()
```

```python
#membandingkan element sekarang dengan element sebelumnya, dikalikan 100. lalu disimpan di dataframe baru
df_per_change = df_adj.pct_change()*100

#membuang baris yang terdapat nilai kosong
df_per_change.dropna(inplace = True, how = "any", axis = 0)

#menampilkan beberapa baris data
df_per_change.head()
```

Dikarenakan dataframe diambil dari dataframe yang telah di-filter tahunnya, sehingga tidak perlu di-filter tahun pada dataframe gabungan ini

###### Visualisasi

```python
volatility = df_per_change[['Nasdaq', 'Hang Seng', 'Nikkei']].rolling(7).std()*np.sqrt(7)
sns.set_theme(context='poster')
sns.relplot(data=volatility,kind="line",height=5, aspect=2.5,dashes = False).set(title="Volatilitas Ketiga Perusahaan")
```

#### Penjelasan 
Dari hasil visualisasi, terlihat bahwa volatilitas perusahaan Nasdaq paling tinggi, sedangkan Nikkei memiliki volatilitas terendah. Hal ini menunjukkan perusahaan Nasdaq dapat menjadi incaran para investor untuk meraup keuntungan lebih tinggi. Sebab Volatilitas yang tinggi menunjukkan potensi profit yang tinggi, namun juga tinggi resiko. Sedangkan Volatilitas rendah menunjukkan rendahnya potensi profit
